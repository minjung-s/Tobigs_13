# 2. 상관관계 확인

heatmap을 통해 모든 연속형 변수 간의 상관관계를 확인할 수 있다. 
seaborn의 함수를 통하여 Heatmap을 생성한다.

상관계수의 절댓값은 0~1 사이인데, 절댓값이 1에 가까울수록 상관관계가 크다고 판단한다.
1에 가까울 수록 양의 상관관계(비례)가 크고, -1에 가까울 수록 음의 상관관계(반비례)가 크다.


```python
sns.heatmap(df.corr(),cmap='summer')
plt.show()
```

위의 Heatmap을 보면, 노란색에 가까울 수록 양의 상관관계가 크고 초록색에 가까울 수록 음의 상관관계가 크다는 것을 알 수 있다. 

양의 상관관계가 큰 예시 : Auction_miscarriage_count&Auction_count  Hammer_price&Minimum_sales_price 등
음의 상관관계가 큰 예시 : point.y&point.x

# 4. 피쳐 생성

### 1
건물의 층수로 저층건물, 준초고층건물, 초고층건물로 나눌 수 있다.
건축법규에 의하면 30층~46층 사이의 건물을 준초고층건물, 50층 이상의 건물을 초고층건물로 규정한다.

```python
df['Total_floor'].plot.hist(bins=10,color='grey',edgecolor='black')
plt.title('Total_floor')
plt.show()
```

```python
def floor(x):
    if 30 <= x < 50:
        return '준초고층건물'
    elif 50 <= x :
        return '초고층건물'
    else:
        return '저층건물'

# 첫 번째 피쳐 생성
df['Total_floor'] = df['Total_floor'].apply(floor) # apply를 이용해 행마다 함수 적용
```

### 2
3. Hammer_price&point.x에서 확인한 결과 사용

```python
plt.scatter(y=df['Hammer_price'],x=df['point.x'])
plt.show()
```

127.5도 이하이면 서울, 그외의 값을 가지면 부산임을 알 수 있다.

```python
def x(x):
    if  x < 127.5:
        return '서울'
    else:
        return '부산'

# 두 번째 피쳐 생성
df['point.x'] = df['point.x'].apply(x) # apply를 이용해 행마다 함수 적용
```

35.5도 이하이면 부산, 그외의 값을 가지면 서울임을 알 수 있다.

### 4
3. Hammer_price&Minimum_sales_price에서 확인한 결과 사용

```python
plt.scatter(x=df['Hammer_price'],y=df['Minimum_sales_price'],label='Minimum_sales_price')
plt.show()
```

위의 그래프에서 최저매각가격이 4,000,000,000이하이면 낙찰가가 4,000,000,000이하임을 알 수 있다.

```python
def min_sales(x):
    if  x < 4e9:
        return '낙찰가 4,000,000,000이하'
    else:
        return '낙찰가 4,000,000,000이상'

# 네 번째 피쳐 생성
df['Minimum_sales_price'] = df['Minimum_sales_price'].apply(min_sales) # apply를 이용해 행마다 함수 적용
```

### 5
3 Hammer_price&Total_appraisal_price에서 확인한 결과 사용

```python
plt.scatter(x=df['Total_appraisal_price'],y=df['Hammer_price'])
plt.show()
```

위의 그래프에서 총감정가가 5,000,000,000이상이면 낙찰가가 4,000,000,000이상임을 알 수 있다.

```python
def total_app_price(x):
    if  x < 5e9:
        return '낙찰가 4,000,000,000이하'
    else:
        return '낙찰가 4,000,000,000이상'

# 다섯 번째 피쳐 생성
df['Total_appraisal_price'] = df['Total_appraisal_price'].apply(total_app_price) # apply를 이용해 행마다 함수 적용
```

채권자가 은행,금융권인 경우와 개인인 경우를 구분할 수 있다.

```python
def log_distribution(x):
    f,ax=plt.subplots(1,2)
    
    x.hist(ax=ax[0],color='black')
    ax[0].set_title(x.name + ' 변환 전')
    
    np.log(x).hist(ax=ax[1],color='green')
    ax[1].set_title(x.name + ' 변환 후')
    plt.show()
```

```python
# 정규분포에 가깝게 변화된 것을 확인
skew=['Auction_count','Hammer_price','addr_bunji1','addr_bunji2']

for cols in skew:
    try :
        log_distribution(df[cols])
    except:
        print(cols +' 는 범위 오버')
```
